```python
# removing other_data entry
df = pd.DataFrame(counties).drop(0).reset_index().drop(columns = 'index');
df.columns = ['Counties']
df
```

```python
df['Year'] = 2016
df = df.sort_values(by ='Counties').reset_index().drop(columns = 'index')
```

```python
df_2 = df.copy(deep = True);
df_2
```

```python
df_2['Year'] = 2012
df_2
```

```python
df = df.merge(df_2, how = 'outer'); 
df
```

```python
#df = df.sort_values(by= 'Counties').reset_index().drop(columns = 'index')
```

```python
winner_dict = {'b': 'Barack Obama', 'h': 'Hillary Clinton', 'm': 'Mitt Romney', 'd': 'Donald Trump'}
pol_winners = []
pol_percent = []
for i in tqdm(range(len(df[91:]))):
  print ('\n\n', df[:91].iloc[i, [0, 1]], '\n')
  print ('Enter election winner:');
  temp_win = input()
  print ('Enter election winner percentage:');
  temp_win_per = input()
  pol_winners.append(winner_dict[temp_win]);
  pol_percent.append(temp_win_per);
  print('\nAdded: ', winner_dict[temp_win], '\n')
```

```python
len(pol_winners)
```

```python
len(df[91:])
```

```python
## added to correct dsirupted loop traversal due to key log error
df_left = df[91:].copy(deep = True)
for i in tqdm(range(len(df_left))):
  print ('\n\n', df_left.iloc[i, [0, 1]], '\n')
  print ('Enter election winner:');
  temp_win = input()
  print ('Enter election winner percentage:');

  temp_win_per = input()
  pol_winners.append(winner_dict[temp_win]);
  pol_percent.append(temp_win_per);
  print('\nAdded: ', winner_dict[temp_win], '\n')
```

```python
df['winning_candidate'] = pol_winners
df['winner_vote_percent'] = pol_percent
```

```python
df
```

```python
entry_SF1 = pd.DataFrame(data=['San Francisco', 2012, winner_dict['b'], 85.3]).transpose()
entry_SF2 = pd.DataFrame(data=['San Francisco', 2016, winner_dict['h'], 85.0]).transpose()
entries_SF = entry_SF1.merge(entry_SF2, how = 'outer');
```

```python
entries_SF.iloc[0, 1] = int(entries_SF.iloc[0, 1])
entries_SF.columns = df.columns
entries_SF
```

```python
df.merge(entries_SF, how = 'outer')
```
