## Select Model and Train

```python
from tqdm import tqdm
```

```python
model = LogisticRegression(multi_class='multinomial', max_iter= 1000)
```

```python
# converting percent strings into raw floating point numbers

for i in tqdm(df.iloc[:, 7:-3].columns):
  df[i] = df[i].apply(lambda x : float(x[:-1]))
df.head()
```

```python
df.iloc[:, 7:-3]
```

```python
X = df.iloc[:, 7:-3]
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
model.fit (X_train, y_train)
```

## Model Evaluation

```python
y_pred = model.predict(X_test);
```

```python
from sklearn.metrics import confusion_matrix, classification_report;
```

```python
print ("Confusion Matrix: ", confusion_matrix(y_test, y_pred), "\n");
print ("Classification Report: ", classification_report(y_test, y_pred), "\n");
```

## Neural networks

```python
import tensorflow as tf
from tensorflow.keras import layers
```

```python
#deep_cols = []
#for i in range(len(X.columns)):
  #deep_cols.append(tf.feature_column.indicator_column(tf.feature_column.numeric_column(X.columns[i])))


model_NN = tf.keras.Sequential([
  #layers.Dense(INPUT_LAYER, activation='relu'),  ### FI ME: Look into this based on data being used
  layers.Dense(512, activation='relu'),
  layers.Dense(256, activation='relu'),
  layers.Dropout(0.2),
  layers.Dense(128, activation='relu'),
  layers.Dense(64, activation='relu'),
  layers.Dense(1, activation='tanh')
])

model_NN.compile(optimizer=tf.keras.optimizers.Adam(),  
              #loss=tf.keras.losses.CategoricalCrossentropy(),
              loss=tf.keras.losses.BinaryCrossentropy(),              
              metrics=['accuracy', 'Recall', 'Precision', 'AUC'])
```

```python
model_NN.fit(x = X_train, y = y_train, epochs = 50, verbose = 2)
```
